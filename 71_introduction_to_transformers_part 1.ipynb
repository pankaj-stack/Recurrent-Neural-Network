- **Origin of Transformer** 
    - 1. pehle encoder-decoder aaya : jisme 30 words se jyada ka sentence agar diya gya to ye fail kr jata tha. q ki ye summary generate krta hai pehle sentence ka phir usko decoder ko bhejta tha. 
    - 2. phir aaya encoder-decoder ko overcome krne ko **attention mechanism** jo ki encoder-decoder se better perform kr rha tha but isme bhi problems the jaise ki ye lstm based encoder-decoder hota hai to isme training word by word hota hai or token by token jisse bare data pe training krne me bahut time lag jaayega ya ye bol sakte ho ki training nhi ho paayegi aur agar training nhi ho paayegi to hum yhan pe **transfer learning** jaisa concept hum use nhi kr paayenge. 
    - 3. aur phir aata hai **Transformer** jisme lstm ko nhi use kiya uski jagah pe **self attention** use kiya gya hai. jisse parallel training hota hai aur phir **transfer learning** ka bhi concept enable ho jata hai NLP ke domain me.
