- lstm ek difficult topic hai little bit
- lstm ka origin isliye hua q ki rnns are not handle long sentences because of vanishing gradient or exploding gradient problem
- simple rnn me short term memory hi reh jati hai baki long term memory vanish ho jati hai. 
- in lstm there are two states in feedback :--> 
    - 1. short term memory 
    - 2. long term memory 
    - jbki rnn me only short term memory hoti hai. 
- lstm me 3-gates hote hain (forget,input and output gates)
- forget gate : current input ke basis pe long term memory me kya remove krna hai ye decide krta hai. 
- input gate  : current input ke basis pe long term memory me kya add krna hai ye decide krta hai.
- output gate : current input ke basis pe long term memory se kya nikal ke output dena hai ye decide krta hai aur short term memory bhi create krta hai.
