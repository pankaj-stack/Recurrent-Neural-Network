```python
# to make range of nos. bw 0 to 1 ,divide the whole nos. by maximum nos. 
max = 100000
for i in range(1,max+1):
    print(f'{i} : {i/max}')
```

- positional encoding is similar to binaary encoding only differece is binary encoding is for descrete nos. while positional encoding for continuous nos.
