- ann -----> works on the tabular data. 
- cnn -----> works on the grid data(images).
- rnn(recurrent neural network) ----> works on the sequential data. 
- rnn is a type of sequential model.

- what is sequential data:
    - any type of textual,time-series,speech(audio) and DNA sequences etc. data is called sequential data.
    - Example : 
        - Hi my name is Nitish.
        - Hello how are you Lord Krishna and Radhe Rani.
